```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
import matplotlib.image as mpimg

%matplotlib inline
```

```python
df = pd.read_csv('searching_k.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
clusters = df.columns[9:]
```

```python
fig, ax = plt.subplots(figsize=(20,20), ncols=1, nrows=len(clusters), squeeze=False)
plt.suptitle("denombrement par cluster", fontsize=16)

for i, x in enumerate(clusters):
    x2=str(x)
    #print(type(x2))
    ax[i][0].set_title('-- {} clusters --'.format(x))
    sns.countplot(x=x2,data=df, ax=ax[i][0])

```

```python
def images_subset_plot(cluster,cat):
    images = []
    subset = df[df[cluster]==cat].sample(20)
    for index, row in enumerate(subset.name):
        images.append(mpimg.imread('../COVER_JPEG/'+row))

    plt.figure(figsize=(20,10))
    plt.suptitle('========{} cluster, cat {} ========'.format(cluster, cat), fontsize=16)
    columns = 5
    for i, image in enumerate(images):
        plt.subplot(len(images) / columns + 1, columns, i + 1)
        plt.imshow(image)
```

```python
images_subset_plot('100',19)
```

```python
images_subset_plot('1000',616)
```
