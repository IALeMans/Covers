```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(color_codes=True)
import matplotlib.image as mpimg

from sklearn.cluster import KMeans

%matplotlib inline
```

```python
df = pd.read_csv('base.csv',sep=';',decimal=',')
df.replace(to_replace=r'properties', value='jpg', regex=True,inplace=True)
df.head()
```

```python
df.describe()
```

```python
X = df.iloc[:, 1:].values
X.shape
```

```python
clusters = [1,3,10,33,100,333,1000,3333,10000,33333]
wcss = []
for cluster in clusters:
    start_time = time.time()
    print('----- processing {} clusters'.format(cluster))
    kmeans = KMeans(n_clusters = cluster, init = 'k-means++', random_state = 0, n_jobs = -1)
    kmeans.fit(X)
    df[cluster] = kmeans.fit_predict(X)
    #print(df.columns)
    df.to_csv('searching_k.csv')
    wcss.append(kmeans.inertia_)
    print('>>>>> {} clusters: elapsed time {}'.format(cluster, time.strftime("%H:%M:%S", time.gmtime(time.time() - start_time))))
```
